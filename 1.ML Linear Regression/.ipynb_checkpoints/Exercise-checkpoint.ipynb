```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def grad(x):
    return 2*x+2
def f(x):
    return x*x+2*x+5
x=5
# learning_rate=0.01
learning_rate = 0.1
for i in range(100):
    if grad(x)<1e-5:break
    else:x=x-learning_rate*grad(x)
# print(x)
print("%.2f"%f(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def grad(x):
    return 2*x+2
def f(x):
    return x*x+2*x+5
x=5
y=[]
for i in range(10):
    x=x-0.1*grad(x)
    y.append(f(x))
plt.plot(y)
plt.xlabel("Số lần")
plt.ylabel("f(x)")
plt.title("Giá trị f(x) sau số lần thực hiện bước 2")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data_square.csv").values
N = data.shape[0]
x = data[:,0].reshape(-1,1)
y = data[:,1].reshape(-1,1)
plt.scatter(x,y)
plt.xlabel("Met vuong")
plt.ylabel("Gia")

x2 = (x[:,0]**2).reshape(-1,1)
x = np.hstack((np.ones((N,1)),x,x2))

# A = np.dot(x.T,x)
# b = np.dot(x.T,y)
# w = np.dot(np.linalg.pinv(A),b)
# print(f"w0 = {w[0]}, w1 = {w[1]}, w2 = {w[2]}")
# print(w)


w = np.array([2000.,-100.,1.]).reshape(-1,1)
numOfIteration = 100
# cost = np.zeros((numOfIteration,1))
learning_rate = 0.0000000001
cost = np.zeros((numOfIteration,1))
for i in range(1,numOfIteration):
    r = np.dot(x,w)-y
    w[0]-=learning_rate*np.sum(r)
    w[1]-=learning_rate*np.sum(np.multiply(r,x[:,1].reshape(-1,1)))
    w[2]-=learning_rate*np.sum(np.multiply(r,x[:,2].reshape(-1,1)))
print(f"w0 = {w[0]}, w1 = {w[1]}, w2 = {w[2]}")

predict = np.dot(x,w)
plt.plot(x[:,1],predict)
plt.show()
```
